```python
def convfechas(x):
    x=x.split('/')
    if int(x[2])<=18:
        return '20'+x[2]+'-'+x[1]+'-'+x[0]
    return '19'+x[2]+'-'+x[1]+'-'+x[0]
```

```python
for i,_ in enumerate(lines):
    lines[i][1]=convfechas(lines[i][1])
lines[:10]
```

```python
def year(x):
    x=x.split('/')
    if int(x[2])<=18:
        return '20'+x[2]
    return '19'+x[2]

def month(x):
    return x.split('/')[1]

def day(x):
    return x.split('/')[0]
```

```python
lines=open('Estaciones.csv','r').readlines()
lines = [x[:-1] if x[-1] == '\n' else x for x in lines]
lines=[x.split(';') for x in lines ]
for i,_ in enumerate(lines):
    lines[i][1]='-'.join([year(lines[i][1]),month(lines[i][1]),day(lines[i][1])])
lines[:10]
```

```python
# Requerimiento 5
lines=open('estaciones.csv.4','r').readlines()
lines = [x[:-1] if x[-1] == '\n' else x for x in lines]
lines = [x.split(';') for x in lines]
lines[:10]

r=[] #resultado final
clave_actual=None
d={}

for row in lines:
    clave =';'.join([row[0],row[3]])
    valor = float(row[9].replace(',','.'))
    
    if clave in d.keys():
        d[clave].append(valor)
    else:
        d[clave] = [valor]

for k in d.keys():
    r.append(k + ':' + str(round((sum(d[k])/len(d[k])),2)))
        
r[:10]
r='\n'.join(r)
open('requerimiento5.csv','w').write(r)
```
